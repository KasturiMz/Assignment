```python
#Q2. Write a python program to print Python Exception Hierarchy.
import inspect
import logging
logging.basicConfig(filename = 'test1.log',level = logging.DEBUG)
def treeclass(cls, ind = 0):
    logging.info("Function created to generate exception hierarchy")
    print('-' * ind, cls.__name__)
    for i in cls.__subclasses__():
        treeclass(i, ind + 3)

print("Hierarchy for exceptions is : ")
logging.info("Printing the hierarchy of exceptions in console")
inspect.getclasstree(inspect.getmro(Exception))
treeclass(Exception)
```

```python
#Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
def tree(cls, ind = 0):
    print('-' * ind, cls.__name__)
    for i in cls.__subclasses__():
        tree(i, ind + 3)

print("Hierarchy for ArithmeticError exceptions is : ")
logging.info("Printing the hierarchy of ArithmeticError exceptions in console")
inspect.getclasstree(inspect.getmro(ArithmeticError))
tree(ArithmeticError)
```
