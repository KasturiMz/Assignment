```python
'''Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
A scatter plot is a visual representation of how two variables relate to each other.
'''
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
'''Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Subplot method provides a way to plot multiple plots on a single figure.
'''
fig, ax = plt.subplots(2,2)
x = np.array([0, 1, 2, 3, 4, 5])
ax[0,0].plot(x,np.array([0, 100, 200, 300, 400, 500]),'b')
ax[0,1].plot(x,np.array([50, 20, 40, 20, 60, 70]),'r')
ax[1,0].plot(x,np.array([10, 20, 30, 40, 50, 60]),'g')
ax[1,1].plot(x,np.array([200, 350, 250, 550, 450, 150]),'k')
fig.show()
```

```python
'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
Barplots are a type of data visualization used to represent data in the form of rectangular bars.
'''
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company vs Profit")
plt.show()
```

```python
plt.barh(company,profit)
plt.ylabel("Company")
plt.xlabel("Profit")
plt.show()
```

```python
'''Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
Boxplots are a measure of how well distributed the data in a data set is. This graph
represents the minimum, maximum, median, first quartile and third quartile in the data set.
'''
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
import pandas as pd
```

```python
data = pd.DataFrame({'box1': box1, 'box2':box2})
ax = data[['box1','box2']].plot(kind = 'box', title = 'boxplot')
plt.show()
```
