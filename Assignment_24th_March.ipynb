```python
df.isnull().sum()
```

```python
df.corr(method='pearson')
```

```python
'''We can handle missing values by various methods:
-We can use mean to replace missing values in datasets with data missing completely at random and no outliers since mean is not robust to outliers.
-We can use median to replace missing values in datasets eith data missing completely at random and outliers present as median is more robust to outliers.
-We can use mode to replace missing values in datasets for categorical variables.
-We can use random values to replace the missing data if the missing data is at random.
-We can use unsupervised clustering methods to replace missing values in datasets with data not missing at random.
'''
```
